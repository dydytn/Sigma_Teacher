#在圓的圓弧上產生波浪

```python
#在圓的圓弧上產生波浪
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10),dpi=80)
pi=3.14159 #圓周率
r1=5
r2=1
s=0
def Draw_Wave(n):  #定義
  t=np.linspace(0,2*pi*n,1001) #(起點,終點,共幾個點包含端點)
  x=r1*np.cos(t/n)
  y=r1*np.sin(t/n)
  plt.plot(x,y,color="red",linewidth=1)

  x=(r1+r2*np.sin(t))*np.cos(t/n)
  y=(r1+r2*np.sin(t))*np.sin(t/n)
  plt.plot(x,y,color="green",linewidth=3)
  
  plt.text(0-r2,0,"n="+str(n),fontsize=15)

for n in [4,5,6,7,8,9,10,20,30]: 
  s=s+1
  plt.subplot(3,3,s)   #作圖s
  Draw_Wave(n)
   
plt.show()
```
